```python
import pandas as pd
import json
```

```python
df = pd.read_csv("./dataset/data-review-indo.csv")
df.head()
```

```python
df_columns = df.columns.tolist()
used_col = ['review_count', 'review_rating', 'reviews_per_rating', 'user_reviews']
```

```python
for i in df_columns:
    if i not in used_col:
        df.drop(i, inplace=True, axis=1)
```

```python
df.head()
```

```python
reviews = df[['user_reviews']]
```

```python
reviews_ratings = []

for i in reviews['user_reviews']:
    if isinstance(i, str):
        place_all_reviews = json.loads(i)
        for people_review in place_all_reviews:
            review_rating = {
                'Rating': people_review.get('Rating', None),
                'Review': people_review.get('Description', '')
            }
            reviews_ratings.append(review_rating)

reviews_ratings_df = pd.DataFrame(reviews_ratings)

reviews_ratings_df.head()
```
